# Hospital Readmissions Data Analysis and Recommendations for Reduction

### Background
In October 2012, the US government's Center for Medicare and Medicaid Services (CMS) began reducing Medicare payments for Inpatient Prospective Payment System hospitals with excess readmissions. Excess readmissions are measured by a ratio, by dividing a hospital’s number of “predicted” 30-day readmissions for heart attack, heart failure, and pneumonia by the number that would be “expected,” based on an average hospital with similar patients. A ratio greater than 1 indicates excess readmissions.

```python
# Generating some variables for a hypothesis test
n_small = len(sub100_ratio[~np.isnan(sub100_ratio)])
n_large = len(abv1k_ratio[~np.isnan(abv1k_ratio)])

mean_small = np.mean(sub100_ratio[~np.isnan(sub100_ratio)])
mean_large = np.mean(abv1k_ratio[~np.isnan(abv1k_ratio)])

std_small = np.std(sub100_ratio[~np.isnan(sub100_ratio)])
std_large = np.std(abv1k_ratio[~np.isnan(abv1k_ratio)])
```

# Formulating an experiment

I am going to start by stating the null and alternative hypothesis.

Null hypothesis: There is no significant difference in the means of readmission ratios for hopsitals with fewer than 100 discharges and hospitals with greater than 1000 discharges.

Alternative hypothesis: There is a significant difference in readmassion rates between the large and small hospitals.

Now that we have determined what the null and alternative hypothesis are, we can do some analysis to determine which hypothesis seems to be correct. We will be using an alpha value of 0.01. We will start by using a two sample t-test as our frequentist approach to this question. 

```python
# Calculating the t statistic
t = (mean_small - mean_large) / np.sqrt((std_small**2 / n_small) + (std_large / n_large))
t
```

```python
from scipy import stats
```

```python
pval = stats.t.sf(np.abs(t), n_large-1)*2
pval
```

# Observing our results
The t statistic and p-value for our experiment were created above. I was not sure on how many degress of freedom to use, but the most common consensus seemed to be the smaller of n-1 between the two samples, which was used. Observing these values, it does seem like our experiment was significant, and there is reason to believe the two hospitals are meaningfully different, since our p-value is low enough to reject our null hypothesis. I will repeat this experiment with bootstrapping, to see if I can obtain a similar result.

# Performing a bootstrap

```python
#importing some functions for bootstrapping
# this function is for drawing permutations of the data
def permutation_sample(data1, data2):
    """Generate a permutation sample from two data sets."""

    # Concatenate the data sets: data
    data = np.concatenate((data1, data2))

    # Permute the concatenated array: permuted_data
    permuted_data = np.random.permutation(data)

    # Split the permuted array into two: perm_sample_1, perm_sample_2
    perm_sample_1 = permuted_data[:len(data1)]
    perm_sample_2 = permuted_data[len(data1):]

    return perm_sample_1, perm_sample_2

# this function draws permutation replicates for a given set of data and function
def draw_perm_reps(data_1, data_2, func, size=1):
    """Generate multiple permutation replicates."""

    # Initialize array of replicates: perm_replicates
    perm_replicates = np.empty(size)

    for i in range(size):
        # Generate permutation sample
        perm_sample_1, perm_sample_2 = permutation_sample(data_1, data_2)

        # Compute the test statistic
        perm_replicates[i] = func(perm_sample_1, perm_sample_2)

    return perm_replicates

# a simple function that calculates the difference of means of two data sets
def diff_of_means(data_1, data_2):
    """Difference in means of two arrays."""

    # The difference of means of data_1, data_2: diff
    diff = np.mean(data_1) - np.mean(data_2)

    return diff
```

```python
# Now to generate our "hacker stats" approach
np.random.seed(115)
# First define our experimental difference of means
small = sub100_ratio[~np.isnan(sub100_ratio)]
large = abv1k_ratio[~np.isnan(abv1k_ratio)]
mean_diff = diff_of_means(small, large)

# Now generate a set of bootstraps to calculate the probability of obtaining this difference of means.
perm_replicates = draw_perm_reps(small, large, diff_of_means, size=10000)

# Compute p-value: p
p = np.sum(abs(perm_replicates) >= mean_diff) / len(perm_replicates)

# Print the result
print('p-value =', p)
```

# Evaluating our bootstrap

So, looking at our bootstrap data, if the means of the two groups were to be the same, we would have never observed what we found in our experiment, so again we find a staticically significant reason to reject our null hypothesis. With this out of the way, I think it is safe to say that there is probably some difference between higher and lower numbers of discharges and the readmission rates.
